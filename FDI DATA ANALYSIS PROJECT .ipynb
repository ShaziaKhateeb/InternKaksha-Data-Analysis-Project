```python
import pandas as pd
from matplotlib import pyplot as plt

# Reading CSV into pandas
data = pd.read_csv(r"C:\Users\shaziya\Desktop\python data analysis p1 data\FSI DATA INTERNATIONAL TRADE top 20.csv")
data.head()
df = pd.DataFrame(data)

name = df['Sector'].head(20)
price = df['2010-11'].head(20)

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(name, price)

# Removing axes splines
for s in ['top', 'bottom', 'left', 'right']:
	ax.spines[s].set_visible(False)

# Removing x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# Showing top values 
ax.invert_yaxis()

# Adding annotation to bars
for i in ax.patches:
	plt.text(i.get_width()+0.2, i.get_y()+0.5, 
			str(round((i.get_width()), 2)),
			fontsize = 10, fontweight ='bold',
			color ='grey')

# Adding Text watermark
fig.text(0.9, 0.15, 'shaziakhateeb', fontsize = 12,
		color ='grey', ha ='right', va ='bottom',
		alpha = 0.7)

# Show Plot
plt.show()
```
